## Playing around with Tesseract

### Images to strings (just words)

```python
print(pytesseract.image_to_string('blurring traditional/Blurring Tradition and Modernity_Judy Van Zile-000001 - 0001.tif'))
```

```python
print(pytesseract.image_to_string(Image.open('blurring tradition/Blurring Tradition and Modernity_Judy Van Zile-000001 - 0001.tif')))
```

### Determining what languages you have installed

```python
print(pytesseract.get_languages(config=''))
```

### OCRing in a non-English language

```python
print(pytesseract.image_to_string('french_text.png', lang='fra'))
```

### Processing txt file of many images

```python
print(pytesseract.image_to_string('images.txt'))
```

### Setting up timeouts for long jobs

```python
try:
    print(pytesseract.image_to_string('test.jpg', timeout=2)) # Timeout after 2 seconds
    print(pytesseract.image_to_string('test.jpg', timeout=0.5)) # Timeout after half a second
except RuntimeError as timeout_error:
    # Tesseract processing is terminated
    pass
```

### Finding confidence for estimates

```python
# Get verbose data including boxes, confidences, line and page numbers
confidence_table= pytesseract.image_to_data('blurring traditional/Blurring Tradition and Modernity_Judy Van Zile-000001 - 0001.tif', output_type='data.frame')
```

## Exporting your content

```python
# Get a searchable PDF
pdf = pytesseract.image_to_pdf_or_hocr('images.txt', extension='pdf')
with open('test.pdf', 'w+b') as f:
    f.write(pdf) # pdf type is bytes by default
```

```python
hocr = pytesseract.image_to_pdf_or_hocr('images.txt', extension='hocr')
```

```python
# Get ALTO XML output
xml = pytesseract.image_to_alto_xml('images.txt')
```
